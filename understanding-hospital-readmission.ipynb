```python
df["readmitted"].value_counts()
```

```python
# Total number of patients
total_patients = len(df)

# Number of patients readmitted within 30 days
readmitted_within_30 = df[df["readmitted"] == "<30"].shape[0]

# Calculate percentage
percentage_within_30 = (readmitted_within_30 / total_patients) * 100

print(f"Percentage of patients readmitted within 30 days: {percentage_within_30:.2f}%")
```

```python
df["age"].value_counts().sort_index()
```

```python
readmitted = df[df["readmitted"].isin(["<30", ">30"])]
```

```python
age_readmit_counts = readmitted["age"].value_counts().sort_index()
age_total_counts = df["age"].value_counts().sort_index()
age_readmit_percent = (age_readmit_counts / age_total_counts) * 100
print(age_readmit_percent)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

age_readmit_percent.sort_index().plot(kind="bar", color="teal")
plt.title("Readmission Percentage by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Readmission Rate (%)")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

```python
df["insulin"].value_counts()
```

```python
insulin_readmit_counts = df.groupby(["insulin", "readmitted"]).size().unstack().fillna(0)
print(insulin_readmit_counts)
```

```python
readmit_under_30 = insulin_readmit_counts["<30"]
total_per_insulin = insulin_readmit_counts.sum(axis=1)

readmit_percent = (readmit_under_30 / total_per_insulin) * 100
print(readmit_percent)
```

```python
readmit_percent.plot(kind="bar", color="tomato")
plt.title("Readmission (<30 days) by Insulin Usage")
plt.xlabel("Insulin Usage Type")
plt.ylabel("Readmission Rate (%)")
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

```python
df["gender"].value_counts()
```

```python
df = df[df["gender"].isin(["Male", "Female"])]
```

```python
gender_readmit = df.groupby(["gender", "readmitted"]).size().unstack().fillna(0)
print(gender_readmit)
```

```python
readmit_gender = (gender_readmit["<30"] / gender_readmit.sum(axis=1)) * 100
print(readmit_gender)
```

```python
readmit_gender.plot(kind="bar", color="skyblue")
plt.title("Readmission Rate (<30 days) by Gender")
plt.xlabel("Gender")
plt.ylabel("Readmission Rate (%)")
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

```python
df['race'].value_counts(dropna=False)
```

```python
df_race = df[df["race"] != "?"]
```

```python
race_readmit= df_race.groupby(["race", "readmitted"]).size().unstack().fillna(0)
readmit_race=(race_readmit["<30"]/race_readmit.sum(axis=1))*100
print(readmit_race)
```

```python
readmit_race.plot(kind="bar", color="orchid")
plt.title("Readmission Rate (<30 days) by Race")
plt.xlabel("Race")
plt.ylabel("Readmission Rate (%)")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

```python
df[["diag_1", "diag_2", "diag_3"]].head()
readmitted_df = df[df["readmitted"] == "<30"]

top_diagnoses = readmitted_df["diag_1"].value_counts().head(10)
print(top_diagnoses)
```

```python
readmitted = df[df["readmitted"] == "<30"]
```

```python
meds = ["metformin", "glipizide", "glyburide", "insulin"]
```

```python
for med in meds:
    med_readmit_counts = readmitted[med].value_counts()
    med_total_counts = df[med].value_counts()

    percent_readmitted = (med_readmit_counts / med_total_counts) * 100
    print(f"\n{med.upper()} Readmission Rates:")
    print(percent_readmitted)
```

```python
metformin_readmit = readmitted["metformin"].value_counts()
metformin_total = df["metformin"].value_counts()

metformin_rate = (metformin_readmit / metformin_total) * 100
metformin_rate = metformin_rate.reindex(["No", "Steady", "Up", "Down"])  # Order nicely

insulin_rate.plot(kind="bar", color="mediumseagreen")
plt.title("Readmission Rate by metformin Usage")
plt.xlabel("metformin Status")
plt.ylabel("Readmission Rate (%)")
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

```python
glipizide_readmit = readmitted["glipizide"].value_counts()
glipizide_total = df["glipizide"].value_counts()

glipizide_rate = (glipizide_readmit / glipizide_total) * 100
glipizide_rate = glipizide_rate.reindex(["No", "Steady", "Up", "Down"])  # Order nicely

glipizide_rate.plot(kind="bar", color="mediumseagreen")
plt.title("Readmission Rate by glipizide Usage")
plt.xlabel("glipizide Status")
plt.ylabel("Readmission Rate (%)")
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

```python
glyburide_readmit = readmitted["glyburide"].value_counts()
glyburide_total = df["glyburide"].value_counts()

glyburide_rate = (glyburide_readmit / glyburide_total) * 100
glyburide_rate = glyburide_rate.reindex(["No", "Steady", "Up", "Down"])  # Order nicely

glipizide_rate.plot(kind="bar", color="mediumseagreen")
plt.title("Readmission Rate by glyburide Usage")
plt.xlabel("glipizide Status")
plt.ylabel("Readmission Rate (%)")
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```
